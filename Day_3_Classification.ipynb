## 8.1 Lojistik Regresyon

Lojistik regresyon temel bir sınıflandırma tekniğidir. Doğrusal sınıflandırıcılar grubuna aittir ve polinom ve doğrusal regresyona biraz benzer. Lojistik regresyon hızlıdır ve nispeten karmaşık değildir ve sonuçları yorumlamanız sizin için uygundur. Temelde bir ikili sınıflandırma yöntemi olmasına rağmen, çok sınıflı problemlere de uygulanabilir.

Classification.png

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
X, y = load_iris(return_X_y=True)
clf = LogisticRegression(random_state=0).fit(X, y)
```

```python
print("Classes: ", clf.classes_)
print("Intercept: ",clf.intercept_)
print("Coef: ",clf.coef_)
```

```python
print(clf.predict_proba(X[:2, :]))
```

```python
print(clf.score(X, y))
```

```python
clf.predict(X[:200, :])
```

Precision ve Recall arasında bir denge aramamız gerekiyorsa VE eşit olmayan bir sınıf dağılımı varsa (çok sayıda Gerçek Negatif) F1 Puanı kullanmak için daha iyi bir ölçü olabilir.

```python
from sklearn.metrics import f1_score, accuracy_score, recall_score, precision_score
```

```python
# 0: Kanser değil
# 1: Kanser

y_true = [0, 1, 1, 0, 0, 0]
y_pred = [0, 0, 1, 0, 0, 1]

print("Accuracy:", accuracy_score(y_true,y_pred))
print("Precision:", precision_score(y_true, y_pred, ))
print("Recall:", recall_score(y_true,y_pred))
print("F1 Score:", f1_score(y_true,y_pred))
```

## 10.1 Aykırı Veri Tespiti ve Veri Ölçekleme (Outlier Detection and Scaling Data)

```python
import numpy as np
```

```python
columns = df.select_dtypes("float64").columns
columns
```

```python
X_new = df.copy()
for column in columns:
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  filter = (X_new[column] >= Q1 - 1.5 * IQR) & (X_new[column] <= Q3 + 1.5 *IQR)
  mean = X_new[column].loc[filter].mean()

  if mean is not np.nan:
    X_new.loc[~filter, column] = mean  # change outlier values with mean
```

```python
  Q1 = df["mean area"].quantile(0.25)
  Q3 = df["mean area"].quantile(0.75)
  IQR = Q3 - Q1
```

```python
Q3 + 1.5 *IQR  # upper limit
```

```python
plt.figure(figsize=(16, 8))
sns.histplot(X_new["mean area"], kde=True)
```

```python
X_new.head()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
X_scaled = MinMaxScaler().fit_transform(X_new)
X_scaled
```

## 10.3 Modeli Eğitmek

```python
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.3, random_state=42)

models = LogisticRegression(random_state=42, n_jobs=-1)
cv = cross_validate(models,X_train,y_train, cv = 3, n_jobs=-1, return_estimator=True)
```

```python
print(f"Mean training accuracy: {np.mean(cv['test_score'])}")
print(f"Test accuracy: { cv['estimator'][0].score(X_test,y_test) }")
```

```python
cv["estimator"]
```

```python
cv['test_score']
```

```python
for i in range(len(cv['estimator'])):
  wrong_answer = 0
  for j in range(len(y_test)):
    result = y_test[j]
    pred = cv['estimator'][i].predict(X_test[j].reshape(1, -1))
    if result != pred:
      wrong_answer += 1

  print(wrong_answer)
```
